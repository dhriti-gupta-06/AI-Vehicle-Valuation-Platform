```python
df_clean = df.copy()
```

```python
df_clean["Engine"] = (df_clean["Engine"].astype(str).str.replace(" cc", "", regex=False))
df_clean["Engine"] = pd.to_numeric(df_clean["Engine"], errors="coerce")
```

```python
df_clean["Engine"].head()
```

```python
df_clean["Max Power"] = (df_clean["Max Power"].str.extract(r'([\d\.]+)'))
df_clean["Max Power"] = pd.to_numeric(df_clean["Max Power"],errors="coerce")
```

```python
df_clean["Max Power"].head()
```

```python
df_clean["Max Torque"] = (df_clean["Max Torque"].str.extract(r'([\d\.]+)'))
df_clean["Max Torque"] = pd.to_numeric(df_clean["Max Torque"],errors="coerce")
```

```python
df_clean["Max Torque"].head()
```

```python
df_clean.info()
```

```python
num_cols = [
    "Engine",
    "Max Power",
    "Max Torque",
    "Length",
    "Width",
    "Height",
    "Seating Capacity",
    "Fuel Tank Capacity"
]
```

```python
for col in num_cols:
    df_clean[col] = (df_clean[col].fillna(df_clean[col].median()))
```

```python
df_clean["Drivetrain"].value_counts()
```

```python
df_clean["Drivetrain"] = (df_clean["Drivetrain"].fillna(df_clean["Drivetrain"].mode()[0]))
```

```python
df_clean.isnull().sum()
```

```python
df_clean.to_csv("clean_vehicle_data.csv",index=False)
```

```python
df_fe = df_clean.copy()
current_year = 2026
df_fe["Vehicle_Age"] = ( current_year - df_fe["Year"])
df_fe[["Year","Vehicle_Age"]].head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_fe["Vehicle_Age"],)
plt.title("Vehicle Age Distribution")
plt.show()
```

```python
df_fe["Power_to_Engine_Ratio"] = (df_fe["Max Power"] / df_fe["Engine"])
```

```python
df_fe[["Engine","Max Power","Power_to_Engine_Ratio"]].head()
```

```python
df_fe["Torque_to_Power_Ratio"] = (df_fe["Max Torque"] / df_fe["Max Power"])
```

```python
df_fe[["Max Torque","Max Power","Torque_to_Power_Ratio"]].head()
```

```python
brand_avg_price = (df_fe.groupby("Make")["Price"].mean())
top_25_per = brand_avg_price.quantile(0.75)
luxury_brands = (brand_avg_price[brand_avg_price >= top_25_per].index.tolist())
```

```python
print(luxury_brands)
```

```python
import numpy as np
```

```python
df_fe["Luxury_Brand"] = np.where(df_fe["Make"].isin(luxury_brands),1,0)
```

```python
df_fe["Luxury_Brand"].value_counts()
```

```python
df_fe["Vehicle_Volume"] = (df_fe["Length"] * df_fe["Width"] * df_fe["Height"])
```

```python
df_fe["Vehicle_Volume"].describe()
```

```python
df_fe["Engine_per_Seat"] = (df_fe["Engine"] / df_fe["Seating Capacity"])
```

```python
df_fe["Engine_per_Seat"].head()
```

```python
df_fe["Automatic_Flag"] = np.where(df_fe["Transmission"]=="Automatic",1,0)
```

```python
feature_eng_cols = [
    "Vehicle_Age",
    "Power_to_Engine_Ratio",
    "Torque_to_Power_Ratio",
    "Luxury_Brand",
    "Vehicle_Volume",
    "Engine_per_Seat",
    "Automatic_Flag",
    "Price"
]
corr = df_fe[feature_eng_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Engineered Features Correlation")
plt.show()
```

```python
X = df_fe.drop("Price", axis=1)
y = df_fe["Price"]
print(X.shape)
print(y.shape)
```

```python
numerical_cols = X.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()
print("Numerical Features:")
print(numerical_cols)
print("\nCategorical Features:")
print(categorical_cols)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import (OneHotEncoder,StandardScaler)
```

```python
numeric_pipeline = Pipeline([("imputer",SimpleImputer(strategy="median")),("scaler",StandardScaler())])
```

```python
categorical_pipeline = Pipeline([("imputer",SimpleImputer(strategy="most_frequent")),("encoder",OneHotEncoder(handle_unknown="ignore"))])
```

```python
preprocessor = ColumnTransformer([("num",numeric_pipeline,numerical_cols),("cat",categorical_pipeline,categorical_cols)])
```

```python
preprocessor
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
print("Training Shape")
print(X_train.shape)
print(y_train.shape)
print("\nTesting Shape")
print(X_test.shape)
print(y_test.shape)
```

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
print(X_train_processed.shape)
print(X_test_processed.shape)
```
