```python
numeric_cols = df.select_dtypes(include=["int64","float64"]).columns
print(numeric_cols)
```

```python
for col in numeric_cols: 
    plt.figure(figsize=(8,6))
    sns.histplot(df[col]) 
    plt.title(col) 
    plt.show()
```

```python
for col in numeric_cols: 
    sns.boxplot(x=df[col] ) 
    plt.title(col)
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include=["str"]).columns
print(categorical_cols)
```

```python
df["Fuel Type"].value_counts()
```

```python
sns.countplot(x="Fuel Type", data=df)
plt.title("Fuel Type Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Transmission", data=df)
plt.title("Transmission Distribution")
plt.show()
```

```python
sns.countplot(x="Owner", data=df)
plt.title("Owner Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(x="Seller Type", data=df)
plt.title("Seller Distribution")
plt.show()
```

```python
sns.countplot(x="Drivetrain", data=df)
plt.title("Drivetrain Distribution")
plt.show()
```

```python
top_brands = df["Make"].value_counts().head(15) 
sns.barplot(x=top_brands.index, y=top_brands.values )
plt.xticks(rotation=90) 
plt.title("Top 15 Brands") 
plt.show()
```

```python
df.groupby("Fuel Type")["Price"].mean().sort_values().plot(kind="bar")
plt.title("Avg Price by Fuel Type")
plt.show()
```

```python
df.groupby("Transmission")["Price"].mean().plot(kind="bar")
plt.title("Avg Price by Transmission")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby("Owner")["Price"].mean().sort_values().plot(kind="bar")
plt.title("Avg Price by Owner Type")
plt.xticks(rotation=20)
plt.show()
```

```python
df.groupby("Year")["Price"].mean().plot(kind="bar")
plt.title("Avg Price by Year")
plt.xticks(rotation=90)
plt.show()
```

```python
corr = df.select_dtypes(include=["int64","float64"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
